# Training a neural network

InÂ thisÂ section,Â weÂ willÂ tryÂ differentÂ hyperparameters,Â oneÂ hyperparameterÂ at a time.Â We willÂ chooseÂ theÂ bestÂ valueÂ of the hypereparameterÂ thatÂ givesÂ usÂ theÂ bestÂ accuracy,Â thenÂ we will takeÂ itÂ andÂ addÂ toÂ itÂ newÂ hyperparameters, andÂ soÂ onÂ until we finishÂ all the hyperparametersÂ thatÂ weÂ use.

```python
# Determine the number of input features
n_features = X_train.shape[1]
print(n_features)
n_output=len(np.unique(y)) # Determine the number of output
print(n_output)
```

```python
from keras.backend import dropout
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras import regularizers

# Function to draw relation between loss and epochs
def history_graph(hist):
  plt.plot(hist.history['loss'])
  plt.plot(hist.history['val_loss'])
  plt.title('Model')
  plt.ylabel('Loss')
  plt.xlabel('Epoch')
  plt.legend('train','val')
  plt.show()
# Function to evaluate the model then return the value of loss and accuracy
def evaluate(model,X,y):
  loss, acc = model.evaluate(X, y)
  return loss, acc
```

## Batch size hyperparameter

InÂ thisÂ section,Â weÂ willÂ try a differentÂ numberÂ ofÂ batchÂ sizesÂ inÂ eachÂ trial.\
Trial 1: batch size = 50\
Trial 2: batch size = 90\
Trial 3: batch size = 6\
WeÂ triedÂ a lot ofÂ values,Â butÂ theseÂ gaveÂ usÂ theÂ largestÂ accuracy.

```python
# Fun to build the model
# num_node parameter is number of neurons in hedden layer
def NN_model(num_node):    
  model=Sequential()
  # hidden layer contains:
  # 1- 192 inputs 
  # 2- 265 neurons in hidden layer
  # 3- tanh as activation function
  model.add(Dense(num_node,activation='tanh' ,input_shape=(n_features,),name='layer1')) 
  # Output layer contains:
  # 1- 99 neurons in output layer
  # 2- softmax as activation function
  model.add(Dense(n_output,activation='softmax' ,name='output'))
  # loss function is a sparse_categorical_crossentropy
  # MeasureÂ theÂ model'sÂ accuracy by accuracy
  model.compile(loss="sparse_categorical_crossentropy",metrics='accuracy')
  return model
```

## RMSprop'sÂ LearningÂ RateÂ OptimizerÂ hyperparameters

InÂ thisÂ section,Â weÂ willÂ try a differentÂ Learning rateÂ ofÂ RMSprop optimizerÂ inÂ eachÂ trial.\

## Regularization

InÂ thisÂ section,Â weÂ willÂ try a differentÂ weight decay for L2 regularizerÂ inÂ eachÂ trial.\
Regularization is usedÂ toÂ preventÂ overfitting,Â soÂ weÂ triedÂ aÂ smallÂ valueÂ ofÂ learningÂ rateÂ asÂ theÂ modelÂ didn'tÂ haveÂ overfitting.\
Trial 1: wd = 0.12\
Trial 2: wd = 0.0013\
Trial 3: wd = 0.0004

```python
# NN_model4 Function to build the model
# It have 3 parameters 
# 1- num_node parameter is number of neurons in hedden layer
# 2- wd is a weight of L2 regularizer
# 3- opLRÂ isÂ the learningÂ rateÂ ofÂ RMSprop'sÂ optimizer.
def NN_model4(num_node, wd, opLR): 
  opt = tf.keras.optimizers.RMSprop(learning_rate=opLR)   # optÂ isÂ aÂ variableÂ thatÂ hasÂ theÂ learningÂ rateÂ ofÂ an RMSprop optimizer.
  model=Sequential()
  # hidden layer contains:
   # 1- 192 inputs 
   # 2- 265 neurons in hidden layer
   # 3- tanh as activation function
   # kernel_regularizer is a L2 regularizer with different weight
  model.add(Dense(num_node,activation='tanh' ,input_shape=(n_features,),name='layer1', kernel_regularizer = regularizers.l2(wd)))
  # Output layer contains:
   # 1- 99 neurons in output layer
   # 2- softmax as activation function
  model.add(Dense(n_output,activation='softmax' ,name='output'))
  # loss function is a sparse_categorical_crossentropy
  # MeasureÂ theÂ model'sÂ accuracy by accuracy
  # optimizerÂ isÂ RMSprop,Â who has a differentÂ learningÂ rate.with learning rate = 0.01
  model.compile(loss="sparse_categorical_crossentropy",metrics='accuracy',optimizer=opt)
  return model
```
